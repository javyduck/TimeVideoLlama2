```python
import json
import numpy as np  # Import numpy for mathematical operations
import random
import math

def radian_to_degree(radian):
    return radian * (180 / math.pi)
# Load human references
with open('video_process/final/conversation_bddx_eval.json', "r") as file:
    list_data_dict = json.load(file)
human_references = [item['conversations'][5]['value'].strip() for item in list_data_dict]

# Load GPT responses
with open('results/BDDX_Test_pred_control_signal.json', 'r') as file:
    data = json.load(file)
    
gpt_responses = [caption['caption'] for caption in data]

def extract_speed_course(text):
    text = text.replace('<t_start>', '').replace('<t_end>', '')
    try:
        parts = text.split()
        speed_index = parts.index('Speed:') + 1
        course_index = parts.index('Course:') + 1
        speed = float(parts[speed_index])
        course = float(parts[course_index])
        return speed, course
    except (ValueError, IndexError) as e:
        print(f"Error processing text: {text} -> {e}")
        return None

human_values = [extract_speed_course(ref) for ref in human_references]
gpt_values = [extract_speed_course(resp) for resp in gpt_responses]
# del human_values[1245]
# del gpt_values[1245]

# Filter None values
filtered_human_values = [value for value in human_values if value is not None]
filtered_gpt_values = [value for value in gpt_values if value is not None]

# Calculate RMSE for Speed and Course
speeds_human, courses_human = zip(*filtered_human_values)
speeds_gpt, courses_gpt = zip(*filtered_gpt_values)

def calculate_rmse(actuals, predictions):
    actuals = np.array(actuals)
    predictions = np.array(predictions)
    return np.sqrt(((actuals - predictions) ** 2).mean())

rmse_speed = calculate_rmse(speeds_human, speeds_gpt)
rmse_course = calculate_rmse(courses_human, courses_gpt)

# Print RMSE results
print(f"RMSE for Speed: {rmse_speed:.2f}")
print(f"RMSE for Course: {rmse_course:.2f}")
```

```python
import json
import numpy as np
from sklearn.metrics import mean_squared_error
# Load human references
with open('video_process/final/normalized_time_conversation_bddx_eval.json', "r") as file:
    list_data_dict = json.load(file)
human_references = [item['conversations'][5]['value'].strip() for item in list_data_dict]

# Load GPT responses
with open('results/finetune_normalized_time_videollama2_test/BDDX_Test_pred_control_signal.json', 'r') as file:
    data = json.load(file)
    
gpt_responses = [caption['caption'] for caption in data ]
def extract_speed_course(text):
    if text.count('<t_start>') != 2 or text.count('<t_end>') != 2:
        return None, None
    text = text.replace('<t_start>', '').replace('<t_end>', '').replace(',', '')
    parts = text.split()
    try:
        speed_index = parts.index('Speed:') + 1
        speed = float(parts[speed_index])
    except:
        speed = None
    try:
        course_index = parts.index('Course:') + 1
        course = float(parts[course_index]) 
    except:
        course = None
    return speed, course

human_values = [extract_speed_course(ref) for ref in human_references]
gpt_values = [extract_speed_course(resp) for resp in gpt_responses]
# Filter None values

# Separate Speed and Course
speeds_human, courses_human = zip(*human_values)
speeds_gpt, courses_gpt = zip(*gpt_values)

# Define sigma thresholds
sigma_values = [0.1, 0.5, 1, 5, 10]

def calculate_metrics(gt_signals, pred_signals, signal_name):
    # Calculate RMSE
    index = np.where(np.array(pred_signals)!=None)
    gt_signals = np.array(gt_signals)[index] 
    pred_signals = np.array(pred_signals)[index] 
    rmse_signal = np.sqrt(mean_squared_error(gt_signals, pred_signals))
    print(f"{signal_name} \t RMSE: {rmse_signal:.2f}")
    
    # Calculate accuracy within sigma thresholds
    for sigma in sigma_values:
        acc = np.count_nonzero(abs(np.array(gt_signals) - np.array(pred_signals)) < sigma) / len(gt_signals)
        print(f"Accuracy within {sigma}: {acc * 100:.2f}")

# Calculate metrics for Speed
calculate_metrics(speeds_human, speeds_gpt, "Speed")

# Calculate metrics for Course
calculate_metrics(courses_human, courses_gpt, "Course")
```

```python
from datasets import load_metric

# Initialize metrics with trust_remote_code to avoid future warnings
bleu = load_metric('bleu', trust_remote_code=True)
meteor = load_metric('meteor', trust_remote_code=True)

results = []
for response, reference in zip(gpt_responses, human_references):
    # Tokenize the prediction for BLEU
    ref = [reference.split()]  # BLEU needs references as a list of lists of tokens
    pred = response.split()    # Prediction tokenized for BLEU

    # Compute BLEU and METEOR scores
    bleu_score = bleu.compute(predictions=[pred], references=[ref])
    meteor_score = meteor.compute(predictions=[response], references=[reference])  # METEOR uses raw strings

    # Store the scores in the results list
    results.append({
        "prediction": response,
        "reference": reference,
        "BLEU": bleu_score['bleu'],
        "METEOR": meteor_score['meteor']
    })

# Calculate mean scores for BLEU and METEOR
mean_bleu = sum(r['BLEU'] for r in results) / len(results)
mean_meteor = sum(r['METEOR'] for r in results) / len(results)

# Print mean scores and check individual entries if needed
print(f"Mean BLEU score: {mean_bleu:.4f}")
print(f"Mean METEOR score: {mean_meteor:.4f}")

# Optionally, print individual results to inspect any anomalies
for result in results:
    print(result)
```
